```python
columns_needed = [
    "POINT_ID",
    "TH_LAT",
    "TH_LONG",
    "LU1",
    "LC1",
    "NUTS0",
    "NUTS2"
]

lucas = lucas_raw[columns_needed].copy()
```

```python
lucas.head()
```

```python
lucas_agri = lucas_raw[
    lucas_raw["LU1"].astype(str).str.startswith("U")
].copy()
```

```python
lucas_agri.shape
lucas_agri["LU1"].value_counts().head(10)
```

```python
lucas_agri["LC1"].value_counts().head(15)
```

```python
lucas_crops = lucas_agri[
    lucas_agri["LC1"].astype(str).str.startswith("B")
].copy()
```

```python
lucas_crops.shape
lucas_crops["LC1"].value_counts().head(10)
```

```python
crop_mapping = {
    # Cereals
    "B11": "wheat",
    "B12": "wheat",
    "B13": "barley",
    "B14": "rye",
    "B15": "oats",
    "B16": "maize",

    # Industrial crops
    "B21": "rapeseed",
    "B22": "sunflower",
    "B23": "soybean",
    "B29": "industrial_other",

    # Root crops
    "B31": "potatoes",
    "B32": "sugar_beet",

    # Fodder & others
    "B41": "fodder",
    "B42": "fodder",

    # Mixed / unclear
    "Bx1": "other_arable",
    "Bx2": "other_arable",
}
```

```python
lucas_crops["crop_class"] = lucas_crops["LC1"].map(crop_mapping)  #Appling the mapping
```

```python
lucas_crops["crop_class"].isna().sum() #Handle unmapped codes
```

```python
#Verify final crop distribution
lucas_crops["crop_class"].value_counts()
```

```python
lucas_phase1_final = lucas_crops[
    [
        "POINT_ID",
        "NUTS0",
        "NUTS2",
        "TH_LAT",
        "TH_LONG",
        "crop_class"
    ]
].copy()
```

```python
lucas_phase1_final.to_csv(
    "lucas2018_DE_phase1_ground_truth.csv",
    index=False
)
```
